```python
import tensorflow as tf
import numpy as np
import tensorflow.keras.layers as layers
from tensorflow.keras.layers import Activation
```

```python
def block(x,i,f):
    
    x = layers.Conv2D(filters=f,kernel_size=3,padding="same", activation="relu")(x)
    if not mini:
        i = i-1
        for i in range(1,i):
            x = layers.Conv2D(filters=f,kernel_size=3,padding="same", activation="relu")(x)
            
    x = layers.MaxPool2D(pool_size=(2,2),strides=(2,2))(x)
 
    return x
        
    
       
def vgg16():
    
    if mini:
        l = [1,1,1,1,1]
    f = [64,128,256,512,512]
    inp = layers.Input((224,224,3))
    
    x =  block(inp,l[0],f[0])
    x =  block(x,l[1],f[1])
    x =  block(x,l[2],f[2])
    x =  block(x,l[3],f[3])
    x =  block(x,l[4],f[4])
    
    x = layers.Flatten()(x)
    
    if Num_class>2:
        activation = 'softmax'
    else:
        activation='sigmoid'
    
    x = layers.Dense(units=4096,activation="relu")(x)
    output = layers.Dense(Num_class, activation=activation, kernel_initializer='he_normal')(x) #multi-class 
    model = tf.keras.Model(inputs=inp, outputs=output, name='Vgg16')
    
    
    return model  
```

```python
mini = True
Num_class = 2
l = [2,2,3,3,3]
model = vgg16()
model.summary()
```
